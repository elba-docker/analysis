```python
# Add analysis to the path
import sys
import pathlib
import os
import numpy

parent_dir = os.path.join(os.path.abspath(''), "..")
sys.path.append(parent_dir)

import parsers
from pprint import pprint

path_to_results = os.path.normpath(os.path.join(parent_dir, "archive"))
print(path_to_results)
working_dir = os.path.normpath(os.path.join(parent_dir, "working"))
data = parsers.main(path_to_results, working_dir=working_dir)
print("Finished")
```

```python
output = os.listdir('/Users/bhanugarg/Documents/school/cs4365/analysis/working')
tot = {}
for i in output:
    temp = i.split(".",1)[0]
    if len(temp) > 0 and "c" in temp:
        nbench_scores = [container.nbench for container in data[temp].replicas[0].single().containers()]
        tot[temp] = {key: numpy.mean([d[key] for d in nbench_scores]) for key in [*nbench_scores[0]]}


for val in tot["i-c-50"]:
    tot["i-mc-50"][val] = (tot["i-mc-50"][val]/tot["i-c-50"][val])*100
    tot["i-rc-50"][val] = (tot["i-rc-50"][val]/tot["i-c-50"][val])*100


c50 = tot["i-c-50"]
mc50 = tot["i-mc-50"]
rc50 = tot["i-rc-50"]
print(c50)
print(mc50)
print(rc50)
    
    

# zipped = zip([i for i in tot if "50" in i],[j for j in tot[0]])
# print(zipped)
# for i in tot:
#     if "50" in i and :
#         print(i)
#         for j in tot[i]:
#             print(str(j) + ": " + str(tot[i][j]))
```
